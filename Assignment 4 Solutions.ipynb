# 1.) Scrape the details of most viewed videos on YouTube from Wikipedia: Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

```python
driver = webdriver.Chrome("chromedriver.exe")
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)

driver.find_element_by_xpath("//b").click()

time.sleep(3)
records = []
for row in driver.find_elements_by_xpath("(//table[@class='wikitable sortable jquery-tablesorter'])[1]//tr")[1:-1]:
    temp = ''
    for element in row.find_elements_by_xpath(".//td"):        
        temp += element.text + '|'
    records.append(temp.split('|')[:-2])

df = pd.DataFrame(records,columns=['rank','video_name','artist','views(M)','date_uploaded'])
    
arranged_cols = ['rank','video_name','artist','date_uploaded','views(M)']
    
df.video_name = df.video_name.apply(lambda x:x[:-4].strip('"'))
df = df[arranged_cols]
df
```

# 2.) Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/.

```python
driver = webdriver.Chrome("chromedriver.exe")
url="https://www.bcci.tv/international/fixtures"
driver.get(url)

Date_Time =[]
Ground=[]
Test_ODI=[]
Team=[]

tm=driver.find_elements_by_xpath("//div[@class='fixture__teams']")
time.sleep(3)
for i in tm:
    if i.text is None :
        Team.append("--") 
    else:
        Team.append(i.text)

G=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
time.sleep(3)
for i in G:
    if i.text is None :
        Ground.append("--") 
    else:
        Ground.append(i.text)
DT=driver.find_elements_by_xpath("//span[@class='fixture__datetime tablet-only']")
time.sleep(3)
for i in DT:
    if i.text is None :
        Date_Time.append("--") 
    else:
        Date_Time.append(i.text)
TO=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/strong")
time.sleep(3)
for i in TO:
    if i.text is None :
        Test_ODI.append("--") 
    else:
        Test_ODI.append(i.text)
```

```python
INDIA_MATCHES =pd.DataFrame([])
INDIA_MATCHES['Team']=Team
INDIA_MATCHES['Date_Time']=Date_Time
INDIA_MATCHES['Series']=Test_ODI
INDIA_MATCHES['Ground']=Ground
INDIA_MATCHES
```

# 4.) Scrape the details of top 100 songs on billboard.com. Url = https://www.billboard.com/

```python
driver = webdriver.Chrome(r"chromedriver.exe")
url="https://www.billboard.com/charts/hot-100"
driver.get(url)
time.sleep(5)

Song_Name =[]
Singer=[]
rank=[]
Last_Week=[]
Weeks_on_board=[]
#scraping the Rank 
rb=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for i in rb:
    if i.text is None :
        rank.append("--") 
    else:
        rank.append(i.text)

#scraping the Song_Name 
son=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
for i in son:
    if i.text is None :
        Song_Name.append("--") 
    else:
        Song_Name.append(i.text)
#scraping the Singer 
sin=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
for i in sin:
    if i.text is None :
        Singer.append("--") 
    else:
        Singer.append(i.text)
#scraping the Last_Week Rank 
lwr=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
for i in lwr:
    if i.text is None :
        Last_Week.append("--") 
    else:
        Last_Week.append(i.text)
#scraping the Weeks_on_board 
wob=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for i in wob:
    if i.text is None :
        Weeks_on_board.append("--") 
    else:
        Weeks_on_board.append(i.text)
BillBOARD_HOT100=pd.DataFrame([])
BillBOARD_HOT100['Peak_rank']=rank
BillBOARD_HOT100['Song_Name']=Song_Name
BillBOARD_HOT100['Singer / Crew']=Singer
BillBOARD_HOT100['Last_Week_Rank']=Last_Week
BillBOARD_HOT100['Weeks_on_board']=Weeks_on_board
BillBOARD_HOT100
```

# 5.) Scrape the details of Highest selling novels. Url = https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-greycompare/

```python
driver = webdriver.Chrome("chromedriver.exe")
url="https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
driver.get(url)
time.sleep(5)

Book_name=[]
Author_name=[]
Volumes_sold=[]
Publisher=[]
Genre=[]
#scraping the Book_name 
bname=driver.find_elements_by_xpath("/html/body/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr/td[2]")
for i in bname:
    if i.text is None :
        Book_name.append("--") 
    else:
        Book_name.append(i.text)
#scraping the Author_name 
Auth=driver.find_elements_by_xpath("/html/body/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr/td[3]")
for i in Auth:
    if i.text is None :
        Author_name.append("--") 
    else:
        Author_name.append(i.text)
#scraping the Genre 
gen=driver.find_elements_by_xpath("/html/body/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr/td[6]")
for i in gen:
    if i.text is None :
        Genre.append("--") 
    else:
        Genre.append(i.text)

#scraping the Publisher 
pub=driver.find_elements_by_xpath("/html/body/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr/td[5]")
for i in pub:
    if i.text is None :
        Publisher.append("--") 
    else:
        Publisher.append(i.text)
#scraping the Volumes_sold 
vs=driver.find_elements_by_xpath("/html/body/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr/td[4]")
for i in vs:
    if i.text is None :
        Volumes_sold.append("--") 
    else:
        Volumes_sold.append(i.text)

Best_Books=pd.DataFrame([])
Best_Books['Book_name']=Book_name
Best_Books['Author_name']=Author_name
Best_Books['Genre']=Genre
Best_Books['Publisher']=Publisher
Best_Books['Volumes_sold']=Volumes_sold
Best_Books
```

# 6.)Scrape the details most watched tv series of all time from imdb.com. Url = https://www.imdb.com/list/ls095964455

```python

driver = webdriver.Chrome("chromedriver.exe")
url="https://www.imdb.com/list/ls095964455"
driver.get(url)
time.sleep(5)
```

```python
Name=[]
Year_span=[]
Genres=[]
Run_time=[]
Ratings=[]
Votes=[]
#scraping the Name 
mname=driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a')
for i in mname:
    if i.text is None :
        Name.append("--") 
    else:
        Name.append(i.text)
#scraping the Year_span 
ys=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
for i in ys:
    if i.text is None :
        Year_span.append("--") 
    else:
        Year_span.append(i.text)
#scraping the Genres 
gnr=driver.find_elements_by_xpath("//span[@class='genre']")
for i in gnr:
    if i.text is None :
        Genres.append("--") 
    else:
        Genres.append(i.text)
#scraping the Run_time 
rt=driver.find_elements_by_xpath("//span[@class='runtime']")
for i in rt[0:101]:
    if i.text is None :
        Run_time.append("--") 
    else:
        Run_time.append(i.text)
#scraping the Ratings 
rate=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]")
for i in rate:
    if i.text is None :
        Ratings.append("--") 
    else:
        Ratings.append(i.text)

#scraping the Votes 
v=driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]/span[2]")
for i in v:
    if i.text is None :
        Votes.append("--") 
    else:
        Votes.append(i.text)
        
IMDB_TOP_100_Series=pd.DataFrame([])
IMDB_TOP_100_Series['Name']=Name
IMDB_TOP_100_Series['Year_span']=Year_span
IMDB_TOP_100_Series['Run_time']=Run_time
IMDB_TOP_100_Series['Genres']=Genres
IMDB_TOP_100_Series['Ratings']=Ratings
IMDB_TOP_100_Series['Votes']=Votes
IMDB_TOP_100_Series
```

# 7.) Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/

```python
driver = webdriver.Chrome("chromedriver.exe")
url="https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(5)
```

```python
search_1 = driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]/a')       # Locating page foe top videos by xpath
search_1.click()
```

```python
Dataset_Name=[]
Data_Type=[]
Task=[]
Attribute_Type=[]
No_of_Instances=[]
No_of_Attribute=[]
Year=[]


#scraping the Dataset_Name 
dname=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
for i in dname:
    if i.text is None :
        Dataset_Name.append("--") 
    else:
        Dataset_Name.append(i.text)
#scraping the Data_Type 
dtype=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
for i in dtype:
    if i.text is None :
        Data_Type.append("--") 
    else:
        Data_Type.append(i.text)

#scraping the Task 
t=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
for i in t:
    if i.text is None :
        Task.append("--") 
    else:
        Task.append(i.text)
#scraping the Attribute_Type 
att=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
for i in att:
    if i.text is None :
        Attribute_Type.append("--") 
    else:
        Attribute_Type.append(i.text)

#scraping the No_of_Instances 
noi=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
for i in noi:
    if i.text is None :
        No_of_Instances.append("--") 
    else:
        No_of_Instances.append(i.text)
#scraping the No_of_Attribute 
noa=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
for i in noa:
    if i.text is None :
        No_of_Attribute.append("--") 
    else:
        No_of_Attribute.append(i.text)
#scraping the Year 
y=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
for i in y:
    if i.text is None :
        Year.append("--") 
    else:
        Year.append(i.text)

UCI=pd.DataFrame([])
UCI['Dataset_Name']=Dataset_Name[:100]
UCI['Data_Type']=Data_Type[:100]
UCI['Task']=Task[:100]
UCI['Attribute_Type']=Attribute_Type[:100]
UCI['No_of_Instances']=No_of_Instances[:100]
UCI['No_of_Attribute']=No_of_Attribute[:100]
UCI['Year']=Year[:100]
UCI
```
